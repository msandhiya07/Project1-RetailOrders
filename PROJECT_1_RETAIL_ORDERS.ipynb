```python
df.rename(columns={'Order Id':'Order_Id','Order Date':'Order_date','Ship Mode':'Ship_mode','Sub Category':'Sub_Category',
                   'Product Id':'Product_Id', 'Postal Code':'Postal_Code','cost price':'Cost_Price',
                   'List Price':'List_Price','Discount Percent':'Discount_Percent'
                   },inplace=True)
df.head()
```

```python

df['Discount'] = df['Cost_Price'] - df['List_price']
df.head()
```

```python

df['Sale_Price'] = df['Cost_Price'] * (1 - df['Discount_Percent'] / 100)
df.head()
```

```python
# Assuming the dataset contains a 'Cost Price' column
df['Profit'] = df['List_price'] - df['Cost_Price']
df.head()
```

```python
df.to_csv('updated_orders.csv', index=False)
df.head()
```

```python
df.to_csv("RetailOrders.csv",index=False)
print("Data saved to 'RetailOrders.csv'")
```

```python
df.to_excel("RetailOrders.xlsx",index=False)
print("Data saved to 'RetailOrders.xlsx'")
```
