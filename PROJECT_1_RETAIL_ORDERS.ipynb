```python
import pandas as pd
import numpy as np
df = pd.read_csv("orders.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python

df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df.to_csv('/content/updated_orders.csv', index=False)

print("Trailing spaces removed and saved as '/content/updated_orders.csv'.")
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
df.head()
```

```python
df.columns.str.lower()
```

```python
df.rename(columns={'Order Id':'order_id','Order_date':'order_date','Ship_mode':'ship_mode','Sub_Category':'sub_category',
                   'Product_Id':'product_id', 'Postal_Code':'postal_code','Cost_Price':'cost_price','Region':'region','State':'state',
                   'List_Price':'list_price','Discount_Percent':'discount_percent','Segment':'segment','Country':'country','City':'city'
                   },inplace=True)
df.head()
```

```python

df['discount'] = df['cost_price'] - df['list_price']
df.head()
```

```python

df['sale_price'] = df['cost_price'] * (1 - df['discount_percent'] / 100)
df.head()
```

```python
# Assuming the dataset contains a 'Cost Price' column
df['profit'] = df['list_price'] - df['cost_price']
df.head()
```

```python
df.to_csv('updated_orders.csv', index=False)
df.head()
```

```python
df.to_csv("Retail_Orders.csv",index=False)
print("Data saved to 'Retail_Orders.csv'")
```

```python
df.to_excel("Retail_Orders.xlsx",index=False)
print("Data saved to 'Retail_Orders.xlsx'")
```
